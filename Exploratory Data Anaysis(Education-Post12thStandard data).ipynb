## Data Visualization

### Univariate Analysis

Lets check the distribution of each variable and presence of outlier using boxplot

```python
df_num=df_org.drop('Names',axis=1)
```

```python
plt.figure(figsize=(12,36))
for i in range(len(df_num.columns)):
    plt.subplot(17,2,(2*i)+1)
    sns.distplot(df_num[df_num.columns[i]])
    plt.title('Distribution of.{}'.format(df_num.columns[i]),weight='bold',color='red',y=1.05)
    plt.subplot(17,2,(2*i)+2)
    sns.boxplot(df_num[df_num.columns[i]])
    plt.title('Boxplot of.{}'.format(df_num.columns[i]),weight='bold',color='Blue',y=1.05)
    plt.tight_layout()
```

**From the above graph,we can observe the following informations** 
* Apps, Accept, Enroll, Top10perc are the variabes which are Right skewed and have many outliers on higher extreme.
* F.Undergrad, P.Undergrad, Personal, Expend are the variabes which are Right skewed and has many outliers on higher extreme.
* Top25perc is the only variable which has no ouliers and closely follows the normal distibution.
* Outstate,Grad.Rate and Room.Board are the variables have very few Outliers and resembles normal distribution.  
* The variable Books has outliers on both the extreme and also right skewed.  
* PhD, Terminal are the variabes are Left skewed and has many outliers on lower extreme.  
* S.F.Ratio have outliers on both extreme and slightly left skewed  
* perc.alumini have few ouliers on higher extreme and slightly left skewed

### Bivariate Analysis

```python
sns.pairplot(df_org)
plt.show()
```

**The above pair plot shows the relationship between the each variable to each other. We can see that there is a co-relation among many of the variables. The value of co-relation has been explained with the aid of heatmap**    

### Heatmap

```python
plt.figure(figsize=(12,12))
sns.heatmap(df_org.corr(),annot=True,mask=np.triu(np.ones_like(df_org.corr(),dtype=np.bool)))
plt.show()
```

**From the above heatmap, we can observe that,**  
* F.Undergrad have strong positive co-relation with variables Apps, Accept and Enroll  
* There is a significant co-relation is exist among many variables.
* The highest negative co-relation is between the variables Expend and S.F.Ratio

**Correlation values near to 1 or -1 are highly positively correlated and highly negatively corelated respectively. Correlation values near to 0 are not correlated to each other.**
